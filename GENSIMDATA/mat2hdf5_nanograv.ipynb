# Convert simulation data from .mat to .hdf5

```python
from os.path import dirname, join as pjoin
import numpy as np
import h5py as h5
import scipy.io as sio

from alive_progress import alive_bar
```

```python
data_dir = '/Users/yiqianqian/Library/Mobile Documents/com~apple~CloudDocs/Research/PulsarTiming/SimDATA/SimNANOGrav'
filename = 'GWBsimDataSKASrlz1Nrlz1'
ext = '.mat'
mat_file = pjoin(data_dir, filename + ext)
mat_file
```

```python
ext = '.hdf5'
h5_file = pjoin(data_dir, filename + ext)
h5_file
```

```python
with h5.File(h5_file,'w') as f:
    f.create_dataset("Np", data=47, dtype="i8")
    with alive_bar(data['psr_data'].shape[1]) as bar:
        for i in range(data['psr_data'].shape[1]):
            psr_data = np.squeeze(data['psr_data'])[i][0][0]
            f.create_group(psr_data[0][0]) # psr name
            f[psr_data[0][0]].create_dataset("alphaP", data=psr_data[1][0])
            f[psr_data[0][0]].create_dataset("deltaP", data=psr_data[2][0])
            f[psr_data[0][0]].create_dataset("yr", data=psr_data[3][0])
            f[psr_data[0][0]].create_dataset("timingResiduals", data=psr_data[5][0])
            f[psr_data[0][0]].create_dataset("sd", data=psr_data[4][0])
            f[psr_data[0][0]].create_dataset("N", data=psr_data[7][0][0], dtype="i8")
            bar()
```
